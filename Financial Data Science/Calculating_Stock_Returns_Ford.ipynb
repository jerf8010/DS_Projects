```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
df = yf.download('F',
                 start ='2018-01-01',
                 end = '2022-05-05'
                 )
df.head()
```

```python
df = df[['Adj Close']].copy()
df.rename(columns = {
    'Adj Close': 'price_t'
}, inplace = True)
df.head()
```

```python
df['returns'] = df['price_t'] / df['price_t'].shift(1) - 1
df.head()
```

```python
df['price_t'].plot(figsize = (12, 8));
```

```python
df['returns'].plot(figsize = (12,8));
```
